```python
img = cv2.imread("test.jpg",1)
resized_img = cv2.resize(img, (10, 10))

added_image = np.clip(img + 250, 0, 255)

# Pengurangan manual (pastikan untuk membatasi nilai minimal 0)
subtracted_image = np.clip(img - 100, 0, 255)

# Perkalian manual (normalisasi dengan membagi 255)
multiplied_image = np.clip((img * 2) / 255, 0, 255)

# Pembagian manual (hindari pembagian dengan nol, normalisasi ke 0-255)
divided_image = np.clip(img / 10, 0, 255)

# Menampilkan hasil operasi
cv2.imshow('Gambar Penjumlahan (NumPy)', added_image.astype(np.uint8))
cv2.imshow('Gambar Pengurangan (NumPy)', subtracted_image.astype(np.uint8))
cv2.imshow('Gambar Perkalian (NumPy)', multiplied_image.astype(np.uint8))
cv2.imshow('Gambar Pembagian (NumPy)', divided_image.astype(np.uint8))

# Tunggu sampai sembarang tombol ditekan untuk menutup jendela
cv2.waitKey(0)

# Menutup semua jendela
cv2.destroyAllWindows()
```

```python
img = cv2.imread("test.jpg",1)
resized_image = cv2.resize(img, (100, 100))

# 1. Translasi (m, n) = (30, 25)
M_translate = np.float32([[1, 0, 30], [0, 1, 25]])  # Matriks translasi
translated_image = cv2.warpAffine(img, M_translate, (img.shape[1], img.shape[0]))

# 2. Rotasi 90 derajat searah jarum jam
rotated_image = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

# 3. Flipping Vertikal
flipped_vertical_image = cv2.flip(img, 0)  # 0 untuk flip vertikal

# 4. Flipping Horizontal
flipped_horizontal_image = cv2.flip(img, 1)  # 1 untuk flip horizontal

# Menampilkan hasil operasi
cv2.imshow('Gambar Asli', img)
cv2.imshow('Gambar Translasi (30, 25)', translated_image)
cv2.imshow('Gambar Rotasi 90 Derajat', rotated_image)
cv2.imshow('Gambar Flipping Vertikal', flipped_vertical_image)
cv2.imshow('Gambar Flipping Horizontal', flipped_horizontal_image)

# Tunggu sampai sembarang tombol ditekan untuk menutup jendela
cv2.waitKey(0)

# Menutup semua jendela
cv2.destroyAllWindows()
```

```python
resized_image = cv2.imread("test.jpg",1)
# resized_image = cv2.resize(img, (10, 10))

# Menghitung histogram untuk setiap channel (B, G, R)
color = ('b', 'g', 'r')
for i, col in enumerate(color):
	hist = cv2.calcHist([resized_image], [i], None, [256], [0, 256])
	plt.plot(hist, color=col)
	plt.xlim([0, 256])

plt.title('Histogram Gambar (10x10)')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')
plt.show()

# Mengonversi gambar ke grayscale untuk ekualisasi histogram
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# Melakukan ekualisasi histogram
equalized_image = cv2.equalizeHist(gray_image)

# Menampilkan gambar asli dan hasil ekualisasi histogram
cv2.imshow('Gambar Asli (10x10)', gray_image)
cv2.imshow('Gambar Hasil Ekualisasi Histogram', equalized_image)

# Menghitung histogram setelah ekualisasi
hist_original = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

# Plot histogram sebelum dan sesudah ekualisasi
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(hist_original, color='gray')
plt.title('Histogram Asli')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')

plt.subplot(1, 2, 2)
plt.plot(hist_equalized, color='gray')
plt.title('Histogram Setelah Ekualisasi')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')

plt.show()

# Tunggu sampai sembarang tombol ditekan untuk menutup jendela
cv2.waitKey(0)
cv2.destroyAllWindows()
```
